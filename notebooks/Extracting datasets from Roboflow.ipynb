```python
from roboflow import Roboflow
import os
import shutil
import uuid
```

```python
# Delete extracted dataset folder
if os.path.exists('/content/images_only'):
    shutil.rmtree('/content/images_only')

# Delete zip file
zip_file = f"/content/images_only.zip"
if os.path.exists(zip_file):
    os.remove(zip_file)
```

```python
# Connect to Roboflow
rf = Roboflow(api_key="UyDdgibxm7iPEzEQfswM")
project = rf.workspace("idrbt").project("vkyc")
version = project.version(2)

# Download dataset
dataset = version.download("yolov8")

# Single folder for ALL images
images_only_dir = "images_only"
os.makedirs(images_only_dir, exist_ok=True)

# Collect images from all splits into ONE folder
for split in ["train/images", "valid/images", "test/images"]:
    split_path = os.path.join(dataset.location, split)
    if os.path.exists(split_path):
        for img in os.listdir(split_path):

            if img.lower().endswith((".jpg", ".jpeg", ".png")):
                new_name = f"{uuid.uuid4().hex}_{img}"

                shutil.copy(
                    os.path.join(split_path, img),
                    os.path.join(images_only_dir, new_name)
                )

print("✅ All images merged into a single 'images_only' folder")


```
